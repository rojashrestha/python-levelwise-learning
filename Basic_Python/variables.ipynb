##Level 3: Multiple Assignment

Assign values 1, 2, 3 to variables x, y, z in one line.

Assign the same value 0 to three variables a, b, and c.

Swap the values of x = 5 and y = 10 without using a third variable.

```python
#Assign values 1, 2, 3 to variables x, y, z in one line.

x,y,z=1,2,3
print(x,y,z)
```

```python
#Assign the same value 0 to three variables a,b,c 

a=b=c=0
print(a,b,c)

```

```python
#Swap the values of x = 5 and y = 10 without using a third variable.

x=5
y=10
x, y =y, x
print("x= ",x )
print("y= ",y)

```
